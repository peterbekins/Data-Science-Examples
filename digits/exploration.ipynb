# Digit Recognizer Project
## Step 2. Data Exploration

The MNIST database includes images of handwritten digits (0-9) with labels. This includes 60,000 images in the train set and 10,000 images in the test set. Each image is stored as a 28x28 array of integers ranging from 0 to 255 to indicate the brightness of each pixel. The images have been centered within the square. 

### 1. Load train data into pandas dfs

I have already converted the raw MNIST data into tidy tables stored as compressed CSV files. These files can now be read into pandas dataframes.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

train_image_file = './data/mnist_train_images.gz'
train_label_file = './data/mnist_train_labels.gz'

x_train = pd.read_csv(train_image_file, index_col = 0, compression = 'gzip')
y_train = pd.read_csv(train_label_file, index_col = 0, compression = 'gzip')

x_train.head()
```

### 2. Examine images by digit

The following function will be used to show the first 15 sample images for each digit as well as a combined image based on the mean value for each pixel.

```python
def plot_images(digit):
    """ This function plots the first 15 images of a given digit from the train set """
    digits = y_train.index[y_train['label'] == digit].tolist()
    print "There are", len(digits), "examples of", digit
    n_col = 15

    plt.figure(figsize=(20,4))
    for i in list(range(n_col)):
        offset =0
        plt.subplot(1, n_col, i + 1)
        grid_data = x_train.iloc[digits[i]].values.reshape(28, 28)  # reshape from 1d to 2d pixel array
        plt.imshow(grid_data, interpolation = "none", cmap = "Greys")
        plt.xticks([])
        plt.yticks([])
    
    plt.figure(figsize=(4,4))
    composite = x_train.iloc[digits].mean().values.reshape(28,28)
    plt.imshow(composite, interpolation = "none", cmap = "jet")
    plt.xticks([])
    plt.yticks([])
    
    plt.show()
```

### 3. Building a composite image

Finally, data from all the images can be combined into a single pattern to illustrate the most commonly filled pixels.

```python
plt.figure(figsize=(4,4))
variance = x_train.var().values.reshape(28,28)
plt.imshow(variance, interpolation = "none", cmap = "jet")
plt.xticks([])
plt.yticks([])
plt.show()
```
